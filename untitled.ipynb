```python
import numpy as np
from keras.models import Model
from sklearn.svm import SVR
from sklearn.linear_model import LinearRegression
from keras.layers import Input, GRU, Dense, Dropout, Lambda, Multiply,Concatenate
from keras import backend as K
import tensorflow as tf
import xgboost as xgb
from sklearn import metrics
from keras.callbacks import EarlyStopping
from keras.regularizers import l2
from keras.optimizers import Adam
```

```python




class CombinedModel:
    def __init__(self, input_shape):
        self.input_shape = input_shape
        self.gru1_model = self.build_gru_model()
        self.xgb1_model = xgb.XGBRegressor(
            n_estimators=1000,
            updater='coord_descent',
            learning_rate=0.5,
            booster='gblinear',
            
            random_state=42,
            
            objective='reg:squarederror' ) # Use Huber loss
        


    def attention_3d_block(self,hidden_states):
     # hidden_states: (batch_size, time_steps, hidden_units)
    
     # Compute the attention scores using Dense layers
     attention_scores = Dense(64, activation='tanh')(hidden_states)  # Shape: (batch_size, time_steps, 64)
     attention_scores = Dense(1, activation='softmax')(attention_scores)  # Shape: (batch_size, time_steps, 1)

     # Reshape attention_scores to (batch_size, time_steps)
     attention_scores = Lambda(lambda x: tf.squeeze(x, axis=-1), output_shape=lambda s: (s[0], s[1]))(attention_scores)  # Shape: (batch_size, time_steps)

     # Apply attention scores by multiplying them with the hidden states
     attention_scores = Lambda(lambda x: tf.expand_dims(x, axis=-1), output_shape=lambda s: (s[0], s[1], 1))(attention_scores)  # Shape: (batch_size, time_steps, 1)
     weighted_states = Multiply()([hidden_states, attention_scores])  # Apply attention weights

     # Sum over the time steps to create a context vector
     context_vector = Lambda(lambda x: tf.reduce_sum(x, axis=1), output_shape=lambda s: (s[0], s[2]))(weighted_states)  # Shape: (batch_size, hidden_units)
    
     return context_vector

    def build_gru_model(self):
     inputs = Input(shape=self.input_shape)

     # Updated GRU layers with 64 hidden units
     gru1 = GRU(64, return_sequences=True, activation='tanh')(inputs)
     dropout1 = Dropout(0.1)(gru1)
    
     gru2 = GRU(64, return_sequences=True, activation='tanh')(dropout1)
     dropout2 = Dropout(0.1)(gru2)
    
     gru3 = GRU(64, activation='tanh',return_sequences=True, kernel_regularizer=l2(0.001))(dropout2)
     dropout3 = Dropout(0.1)(gru3)

     # Apply attention mechanism on the combined GRU output
     attention_output = self.attention_3d_block(dropout3)  # Shape: (None, 64)
     print("attention  output shape",attention_output.shape)
     # Final dense layer with 64 units and sigmoid activation
     gru_output = Dense(1,name="gru_dense_output")(attention_output)
                                                                        # Shape: (None, 64)
     
     print("gru output shape",gru_output.shape)
     print("y_train shape",y_train.shape)
     # Create model
     model = Model(inputs=inputs, outputs=gru_output)
     model.compile(optimizer='adam', loss='mse')  # Updated learning rate
     print(model.summary())
     return model

    
    def fit(self, X_train, y_train,cols):
     import pandas as pd


     features_to_exclude = [
    'close_price_lag_1', 'compound_lag_1', 'open_price_lag_1', 'high_lag_1', 'low_lag_1', 'volume_lag_1',
    'EMA_10_lag_1', 'RSI_lag_1', 'SMA_5_lag_1', 'SMA_10_lag_1', 'SMA_15_lag_1', 'SMA_30_lag_1',
    'close_price_lag_7', 'compound_lag_7', 'open_price_lag_7', 'high_lag_7', 'low_lag_7', 'volume_lag_7',
    'EMA_10_lag_7', 'RSI_lag_7', 'SMA_5_lag_7', 'SMA_10_lag_7', 'SMA_15_lag_7', 'SMA_30_lag_7',
    'close_price_rolling_mean_7', 'compound_rolling_mean_7', 'open_price_rolling_mean_7', 'high_rolling_mean_7', 'low_rolling_mean_7', 
    'volume_rolling_mean_7', 'EMA_10_rolling_mean_7', 'RSI_rolling_mean_7', 'SMA_5_rolling_mean_7', 
    'SMA_10_rolling_mean_7', 'SMA_15_rolling_mean_7', 'SMA_30_rolling_mean_7', 'close_price_rolling_mean_14',
    'compound_rolling_mean_14', 'open_price_rolling_mean_14', 'high_rolling_mean_14', 'low_rolling_mean_14',
    'volume_rolling_mean_14', 'EMA_10_rolling_mean_14', 'RSI_rolling_mean_14', 'SMA_5_rolling_mean_14',
    'SMA_10_rolling_mean_14', 'SMA_15_rolling_mean_14', 'SMA_30_rolling_mean_14', 'close_price_rolling_std_7',
    'compound_rolling_std_7', 'open_price_rolling_std_7', 'high_rolling_std_7', 'low_rolling_std_7',
    'volume_rolling_std_7', 'EMA_10_rolling_std_7', 'RSI_rolling_std_7', 'SMA_5_rolling_std_7', 
    'SMA_10_rolling_std_7', 'SMA_15_rolling_std_7', 'SMA_30_rolling_std_7', 'close_price_rolling_std_14',
    'compound_rolling_std_14', 'open_price_rolling_std_14', 'high_rolling_std_14', 'low_rolling_std_14',
    'volume_rolling_std_14', 'EMA_10_rolling_std_14', 'RSI_rolling_std_14', 'SMA_5_rolling_std_14',
    'SMA_10_rolling_std_14', 'SMA_15_rolling_std_14', 'SMA_30_rolling_std_14', 'close_price_fft_magnitude',
    'compound_fft_magnitude', 'open_price_fft_magnitude', 'high_fft_magnitude', 'low_fft_magnitude', 
    'volume_fft_magnitude', 'EMA_10_fft_magnitude', 'RSI_fft_magnitude', 'SMA_5_fft_magnitude', 
    'SMA_10_fft_magnitude', 'SMA_15_fft_magnitude', 'SMA_30_fft_magnitude'
     ]
     X_train = X_train.reshape(X_train.shape[0], X_train.shape[1])
     X_train_df = pd.DataFrame(X_train, columns=cols)

     # Assuming you have a DataFrame 'X_train' and 'X_test' for your features
     X_train_filtered = X_train_df.drop(columns=features_to_exclude, errors='ignore')
     
     X_train_filtered = X_train_filtered.values
     
     X_train_filtered = X_train_filtered.reshape(X_train_filtered.shape+(1,))
     
     # Fit GRU model
     self.gru1_model.fit(X_train_filtered, y_train, validation_split=0.2, epochs=10, verbose=1,batch_size=16)

     # Get predictions from the GRU model
     gru_predictions = self.gru1_model.predict(X_train_filtered)
     
     features_to_excludexg = ['close_price','open_price','high','low','volume','compound','compound_shifted']
     X_train_filteredxg = X_train_df.drop(columns=features_to_excludexg, errors='ignore')
     # Reshape X_train and X_test from (samples, 7, 1) to (samples, 7)
     X_train_filteredxg = X_train_filteredxg.values
     X_train_reshaped = X_train_filteredxg.reshape(X_train_filteredxg.shape[0], X_train_filteredxg.shape[1])
     gru_predictions_df = pd.DataFrame(gru_predictions, columns=[f"gru_prediction_{i}" for i in range(gru_predictions.shape[1])])
     
     X_train_combined = pd.concat(
     [pd.DataFrame(X_train_reshaped, columns=[f"feature_{i}" for i in range(X_train_reshaped.shape[1])]),
     gru_predictions_df],
     axis=1)
     
     
     

     
      
     # Fit XGBoost model with combined training data
     self.xgb1_model.fit(X_train_combined,y_train)


    def predict(self, X_test,cols):
        features_to_exclude = [
    'close_price_lag_1', 'compound_lag_1', 'open_price_lag_1', 'high_lag_1', 'low_lag_1', 'volume_lag_1',
    'EMA_10_lag_1', 'RSI_lag_1', 'SMA_5_lag_1', 'SMA_10_lag_1', 'SMA_15_lag_1', 'SMA_30_lag_1',
    'close_price_lag_7', 'compound_lag_7', 'open_price_lag_7', 'high_lag_7', 'low_lag_7', 'volume_lag_7',
    'EMA_10_lag_7', 'RSI_lag_7', 'SMA_5_lag_7', 'SMA_10_lag_7', 'SMA_15_lag_7', 'SMA_30_lag_7',
    'close_price_rolling_mean_7', 'compound_rolling_mean_7', 'open_price_rolling_mean_7', 'high_rolling_mean_7', 'low_rolling_mean_7', 
    'volume_rolling_mean_7', 'EMA_10_rolling_mean_7', 'RSI_rolling_mean_7', 'SMA_5_rolling_mean_7', 
    'SMA_10_rolling_mean_7', 'SMA_15_rolling_mean_7', 'SMA_30_rolling_mean_7', 'close_price_rolling_mean_14',
    'compound_rolling_mean_14', 'open_price_rolling_mean_14', 'high_rolling_mean_14', 'low_rolling_mean_14',
    'volume_rolling_mean_14', 'EMA_10_rolling_mean_14', 'RSI_rolling_mean_14', 'SMA_5_rolling_mean_14',
    'SMA_10_rolling_mean_14', 'SMA_15_rolling_mean_14', 'SMA_30_rolling_mean_14', 'close_price_rolling_std_7',
    'compound_rolling_std_7', 'open_price_rolling_std_7', 'high_rolling_std_7', 'low_rolling_std_7',
    'volume_rolling_std_7', 'EMA_10_rolling_std_7', 'RSI_rolling_std_7', 'SMA_5_rolling_std_7', 
    'SMA_10_rolling_std_7', 'SMA_15_rolling_std_7', 'SMA_30_rolling_std_7', 'close_price_rolling_std_14',
    'compound_rolling_std_14', 'open_price_rolling_std_14', 'high_rolling_std_14', 'low_rolling_std_14',
    'volume_rolling_std_14', 'EMA_10_rolling_std_14', 'RSI_rolling_std_14', 'SMA_5_rolling_std_14',
    'SMA_10_rolling_std_14', 'SMA_15_rolling_std_14', 'SMA_30_rolling_std_14', 'close_price_fft_magnitude',
    'compound_fft_magnitude', 'open_price_fft_magnitude', 'high_fft_magnitude', 'low_fft_magnitude', 
    'volume_fft_magnitude', 'EMA_10_fft_magnitude', 'RSI_fft_magnitude', 'SMA_5_fft_magnitude', 
    'SMA_10_fft_magnitude', 'SMA_15_fft_magnitude', 'SMA_30_fft_magnitude'
     ]
        X_test = X_test.reshape(X_test.shape[0], X_test.shape[1])
        X_test_df = pd.DataFrame(X_test, columns=cols)
        X_test_filtered = X_test_df.drop(columns=features_to_exclude, errors='ignore')
        X_test_filtered = X_test_filtered.values
        X_test_filtered = X_test_filtered.reshape(X_test_filtered.shape+(1,))
        gru_predictions=self.gru1_model.predict(X_test_filtered)
        
        plt.figure(figsize=(10, 5))
        for i in range(1):
         plt.plot(gru_predictions[:, i], label=f'Feature {i}')
         plt.plot(y_test, label='Target Price', color='black', linewidth=1)  # Add target price for comparison
         plt.title('All GRU Features vs. Target Price')
         plt.xlabel('Time')
         plt.ylabel('Value')
         plt.legend()
         plt.show()
        print(gru_predictions)
        features_to_excludexg = ['close_price','open_price','high','low','volume','compound','compound_shifted']
        X_test_filteredxg = X_test_df.drop(columns=features_to_excludexg, errors='ignore')
        X_test_filteredxg = X_test_filteredxg.values
        X_test_reshaped = X_test_filteredxg.reshape(X_test_filteredxg.shape[0], X_test_filteredxg.shape[1])
        gru_predictions_df = pd.DataFrame(gru_predictions, columns=[f"gru_prediction_{i}" for i in range(gru_predictions.shape[1])])
       
        X_test_combined = pd.concat(
        [pd.DataFrame(X_test_reshaped, columns=[f"feature_{i}" for i in range(X_test_reshaped.shape[1])]),
        gru_predictions_df],
        axis=1)
        


        # Predict using the XGBoost model
        xgb1_predictions = self.xgb1_model.predict(X_test_combined)

       

        
        return xgb1_predictions
```
